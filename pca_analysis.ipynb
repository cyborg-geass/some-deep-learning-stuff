```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
%matplotlib inline 
```

```python
from sklearn.datasets import load_breast_cancer
dataset1 = load_breast_cancer()
dataset1.keys()
```

```python
dataset1['DESCR']
```

```python
dataset1['target_names']
```

```python
dataset1['feature_names']
```

```python
dataset1['target']
```

```python
df = pd.DataFrame(dataset1['data'], columns=dataset1['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
scaled_data = std_scale.fit_transform(df)
scaled_data
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(20,12))
plt.scatter(x_pca[:,0],x_pca[:,1],c=dataset1['target'],cmap="plasma", alpha=0.9, s=100)
plt.xlabel("1st principal component")
plt.ylabel("2nd principal component")
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=dataset1['feature_names'])
plt.figure(figsize=(20,15))
sns.heatmap(df_comp, cmap="plasma")
```
